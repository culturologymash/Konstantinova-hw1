39. (2 балла) Создайте словарь, в котором ключ -- слово из текста, а значение -- количество раз, сколько оно встречается во всем тексте.

например:

`dct = {"go": 1, "went": 2, "gone": 3}`

Напечатайте словарь

````python
m = [[n.strip(punctuation).strip("«»")  for n in i] for i in s]
u = sum(m,[])
S = dict((l, u.count(l)) for l in set(u))
print("словарь: ", S)

p = sorted(S.values(), reverse = True)
e = {}

for q in p:
  for j in S.keys():
    if S[j] == q:
      e[j] = S[j]
print('словарь,где ключи идут от большего к меньшему:', e)
````
